```python
from IPython.display import Image
import os
import random

im_list = os.listdir('logos/')

names_list = []
for value in im_list:
    names_list.append(value[0:-4])
    
cities_file = open("cities.txt", "r")
cities_file_content = cities_file.read()
cities_list = cities_file_content.split(",")
cities_file.close()

qb_file = open("quarterbacks.txt", "r")
qb_file_content = qb_file.read()
qb_list = qb_file_content.split(",")
qb_file.close()
```

```python
from IPython.display import Image 

index = random.randint(0, len(cities_list))

print('Team Logo: ')
print('\n')
im = Image(filename=("logos/" + im_list[index]), width=200, height=200)
display(im)
print('Team Name: {0}'.format(names_list[index]))
print('\n')
print('City: {0}'.format(cities_list[index]))
print('\n')
print('Quarterback: {0}'.format(qb_list[index]))
```
